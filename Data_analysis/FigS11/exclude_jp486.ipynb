```python
# Make a figure with just the top two rows and save them to plot for latter
GLOBALFONTSIZE = 10
from matplotlib.image import imread
exampleNumber = 3

nrows = 5
ncols = 4
colSize = 2.38
rowSize = 2.2
sns.set(rc={"figure.dpi":300, 'savefig.dpi':300})
sns.set_theme(style = 'ticks')

def add_text(ax,x,y,label,fw = 'normal'):
    ax.text(x,y,label, style='normal', fontweight=fw, fontsize=GLOBALFONTSIZE+3,
                  verticalalignment='center', horizontalalignment='center', transform=ax.transAxes)
    
boxPlotColors = ['#ffecd1','#ff7d00','#15616d','#78290f'] #Lever First,

fig = plt.figure(figsize=(ncols*colSize,nrows*rowSize),constrained_layout=True) 
gs = gridspec.GridSpec(nrows, ncols, figure=fig,width_ratios=[1,1,1,1.1]) 

ax = fig.add_subplot(gs[:2, :])
fn= myProject.dataPath + "/results/anchoring_exclude_jp486_top.png"
img = imread(fn)
ax.imshow(img)
ax.axis('off')

ax = fig.add_subplot(gs[2:, :3])
fn= myProject.dataPath + "/results/anchoring_exclude_jp486_bottom.png"
img = imread(fn)
ax.imshow(img)
ax.axis('off')

gs6 = gridspec.GridSpecFromSubplotSpec(1,1,subplot_spec= gs[:,:])
ax0 = fig.add_subplot(gs6[:,:])
# # #Add figure legend
add_text(ax0,0,1,'a',fw='bold')
add_text(ax0,0.27,1,'b',fw='bold')
add_text(ax0,0.52,1,'c',fw='bold')

add_text(ax0,0,0.8,'d',fw='bold')

add_text(ax0,0.78,1,'e',fw='bold')
add_text(ax0,0.78,0.8,'f',fw='bold')

add_text(ax0,0,0.555,'g',fw='bold')
add_text(ax0,0.27,0.555,'h',fw='bold')
add_text(ax0,0.52,0.555,'i',fw='bold')

add_text(ax0,0,0.34,'j',fw='bold')

ax0.axis('off')

plt.savefig('../../Output/Sup_Figure11.pdf', bbox_inches='tight', dpi=300)
```
